The edge detection can be achieved through the topology theroem. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('naturo-monkey-selfie.jpg')
edges = cv2.Canny(img, 200, 300)
img2 = cv2.imread('house.jpg')
edges2 = cv2.Canny(img2, 100, 200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

Here is a comparision of two different types of images. Clearly we can see that the edge detection result is much better with the image of a house. The therom is to use intensity variation to define the slope and direction of a surface through its gradient. However, we can see that with a image like the monkey's face we accentuates high frequencies since the brightness intensity is high within the facail contour. Therefore, to concur that issue, we could apply a low pass filter first then do the edge detection. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('naturo-monkey-selfie.jpg')

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

Here we first apply the low pass filter using the cv2.filter function.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('naturo-monkey-selfie.jpg')

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
edges = cv2.Canny(dst, 100, 200)

plt.subplot(121),plt.imshow(dst),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

Then we can see that the image contour is much clearer after the image got blurred down.
