```python
from sklearn.metrics import roc_curve
import numpy as np
import PIL
from PIL import Image
import scipy.io as io
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random

im = Image.open('../data/characters.png','r')

imm = im.convert('L')
image = np.array(imm)
print(image.shape)
image = image.astype(np.uint8)
pat = io.loadmat('../data/template-h.mat')
pat = pat['T']





def sp_noise(image,prob):
    
    shape = np.shape(image)
    output = np.zeros(shape,np.uint8)
    thres = 1 - prob 
    for i in range(shape[0]):
        for j in range(shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output
    
    
noise_img = sp_noise(image,0.05)   
noise_img
result = Image.fromarray(noise_img)
result

```

```python
res = np.zeros((noise_img.shape[0] - pat.shape[0], noise_img.shape[1] - pat.shape[1]))
for i in range(res.shape[0]):
    for j in range(res.shape[1]):
        res[i][j] = np.sum((noise_img[i:i+pat.shape[0], j:j+pat.shape[1]]/255-pat)**2)
        
thresh = 30

for i in range(res.shape[0]):
    for j in range(res.shape[1]):
        if res[i, j] < thresh:
#image[i:i + 2, j: j+2] = 0
            noise_img[i:i + pat.shape[0], j] = 0
            noise_img[i:i + pat.shape[0], j + pat.shape[1]] = 0
            noise_img[i, j:j + pat.shape[1]] = 0
            noise_img[i + pat.shape[0], j:j + pat.shape[1]] = 0
            
result = Image.fromarray(noise_img)
result           
```
