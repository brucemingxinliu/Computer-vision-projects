Low pass filter acts on a blurring effect on the original image. In contrast to how much the original image is being blurred, it actually applies better on the noises on the image. So in this case, you can apply both low-pass and high-pass filters on the image to decrease its noises and sharpens the image. 

```python
import numpy as np
import scipy.io as sc
import PIL 
from PIL import Image
import matplotlib
from matplotlib import pyplot as plt
from scipy.fftpack import dct, idct


img1= Image.open('lena.png')
img2= Image.open('naturo-monkey-selfie.jpg')
img3= Image.open('view.jpg')

img1
```

Here are the three images I'm using for the purpose of this task. 

```python
img_matrix= Image.open('lena.png')
img = img_matrix.convert('L')
IMG = np.array(img)
tt = dct(dct(IMG.T,  norm='ortho').T, norm = 'ortho')

tt[40:,40:] = 0
low_filter = idct(idct(tt.T,  norm='ortho').T, norm = 'ortho')

newimg = (IMG - low_filter)
sec = IMG - newimg
plt.imshow(sec, cmap='gray')
plt.show()
```

```python
tt[10:,10:] = 0
low_filter = idct(idct(tt.T,  norm='ortho').T, norm = 'ortho')

newimg = abs(IMG - low_filter)
sec = IMG - newimg
plt.imshow(newimg, cmap='gray')
plt.show()
```
