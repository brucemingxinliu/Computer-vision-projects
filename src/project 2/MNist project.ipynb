MINIST is a database of handwritten digits used to train the computer to recognize handwritting digits using a neural network. But here for the project's purpose, we are using PCA (principal component analysis) to apply the MNIST. 

PCA can be seen as dataset fitting an n-dimensional ellipsoid, which each axis of the ellipsoid represents a principal component. The size of the principal component also shows the size of the covariance. 

To find the axes, we need to subtract the mean of each variable from the dataset to center the data around the origin, then compute the cavariance matrix of the data.
